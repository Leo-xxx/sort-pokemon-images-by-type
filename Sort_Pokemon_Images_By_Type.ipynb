```python
# Move image files into corresponding directories based on Pokemon type
for image_path in glob.glob('pokemon/*.*'):
    pokemon_name = image_path.split('/')[1].split('.')[0]
    print(pokemon_name)
    
    # Get pokemon type from dataframe
    pokemon_type = df.loc[df['Name'] == pokemon_name].iloc[0]['Type1']
    dir_name = pokemon_type.lower()
    
    new_path = 'pokemon/' + dir_name + '/' + pokemon_name + '.' + 'png'
    print('New path: ', new_path)
    
    shutil.copyfile(image_path, new_path)
```
